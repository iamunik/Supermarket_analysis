```python
# Importing the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


t_data = 'supermarket_sales.xlsx'
data = pd.read_excel(t_data)
data.head()
```

```python
# Removing the spaces in the column name and replacing it with '_'
data.columns = [column.replace(" ", "_") for column in data.columns]
```

```python
data.info()
```

```python
data.describe()
```

## KPI's

```python
# Total Revenue
Total_revenue = data['Total'].sum()
print("The total revnue is -> $", format(Total_revenue, ",.2f"))
```

```python
# Total cost

Total_cost = data['cogs'].sum()
print("The total costs of sale is -> $", format(Total_cost, ",.2f"))
```

```python
# Gross profit

data["gross_profit"] = data.Total - data.cogs
Gross_profit = data["gross_profit"].sum()
print("The total gross profit is -> $", format(Gross_profit, ",.2f"))
```

```python
# Average rating

Average_rating = data['Rating'].mean()
print(f"The average rating is -> {round(Average_rating, 2)}")
```

```python
# Total counts of sales

counts = len(data.Invoice_ID.unique())
print(f"We have a total of {counts} sales performed")
```

## Trend Analysis

```python
# Added a 'Month' column

data['Month'] = data['Date'].dt.strftime('%m')
```

```python
# Grouped this based on month for appropriate visualisation

tr = ['Total', 'cogs', 'Quantity', 'Month']
month = data[tr].groupby(['Month']).sum()
month
```

```python
ax = month["Total"].plot(grid=True, marker='.', markersize=10, title='Trend in Revenue based on month', figsize=(10, 5))
ax.set(xlabel='Month', ylabel='Total Revenue')
plt.show()
```

<b>This particular plot shows that the month of January generated the highest revenue compared to the remaining months</b>

```python
tr_i = ['Total', 'cogs', 'Quantity', 'Time']
time = data[tr_i].groupby(['Time']).sum()
time
```

```python
ax = time["Total"].plot(grid=True, marker='.', markersize=5, title='Trend in Revenue based on time', figsize=(20, 5))
ax.set(xlabel='Month', ylabel='Price')
plt.show()
```

```python
c = time.nlargest(1, 'cogs')
print(f"The time that had the highest sales is '{c.Total.index[0]}'")
c
```

## REVENUE ANALYSIS

```python
tr_b = ['Total', 'Branch']
branch = data[tr_b].groupby(['Branch']).sum()
branch
```

```python
branch = branch.sort_values('Total')
branch.plot(kind='barh', grid=True, figsize=(10, 5), title='Revenue analysis based on Branches')
plt.show()
```

<b>Branch C generated the highest revenue compared to other branches of the company</b>

```python
feat = ['Total', 'Gender']
gender = data[feat].groupby(['Gender']).sum()
print(data['Gender'].value_counts())
gender
```

```python
gender.plot(kind="bar", grid=True, title="Revenue analysis based on Gender")
plt.show()
```

<b>Based on the chart above we can see that the Female paying customers produced brought in more revenue than the male customers</b>

```python
tr_c = ['Total', 'Customer_type']
c_type = data[tr_c].groupby(['Customer_type']).sum()
print(data['Customer_type'].value_counts())
c_type
```

```python
c_type.plot(kind='barh', grid=True,figsize=(9, 2), title="Revenue analysis based on Customer Type")
plt.show()
```

<b>Members bring in more revenues to the business than normal customers</b>

```python
n_lst = ['Total', 'Customer_type', 'Gender']
n_type = data[n_lst].groupby(['Customer_type', "Gender"]).sum()
n_type
```

```python
n_type.plot(kind='barh', grid=True,figsize=(9, 2), title="Revenue analysis based on Customer Type and Gender")
plt.show()
```

<b>Gender Analysis showed that Ladies brought more in revenue than the men but this analysis showed that Normal Male customers brought the company more than the men that are Members</b>

```python
tr_p = ['Total', 'Product_line']
p_l = data[tr_p].groupby(['Product_line']).sum()
p_l
```

```python
p_l = p_l.sort_values('Total')
p_l.plot(kind='pie', grid=True,
         figsize=(10,5), subplots=True,
         autopct='%1.1f%%', labels=None,
         pctdistance=1.2, wedgeprops={"edgecolor" : "white", 'linewidth': 1, 'antialiased': True}
         )
plt.axis('equal')
plt.legend(p_l.index, loc='center')
plt.title("Revenue analysis based on Product line", y=1.1)
plt.show()
```

<b>'Food and beverages' brought in more in revenue than other product lines<b>

```python
tr_pr = ['Total', 'Payment']
p_ = data[tr_pr].groupby(['Payment']).sum()
p_
```

```python
p_ = p_.sort_values('Total')
p_.plot(kind='barh', grid=True, title="Revenue analysis based on Payment")
plt.show()
```

<b>People preferred paying cash compard to other method of payments hence the reason why cash is generating more income than other payment types</b>

## Cost Analysis

```python
cst = ['cogs', 'Product_line']
c_p = data[cst].groupby(['Product_line']).sum()
c_p
```

```python
c_p = c_p.sort_values('cogs')
c_p.plot(kind='barh', grid=True, title="Cost of goods analysis based on Payment")
plt.show()
```

<b>'Food and beverages' costs more in production in comparison to other product lines<b>

```python
cst_b = ['cogs', 'Branch']
c_b = data[cst_b].groupby(['Branch']).sum()
c_b
```

```python
c_b = c_b.sort_values('cogs')
c_b.plot(kind='barh', grid=True, title="Cost of goods analysis based on Branches")
plt.show()
```

<b>The costs of goods sold at Branch C is more than the costs of goods sold in other Branches</b>

## RATINGS ANALYSIS

```python
rate_pa = ['Payment', 'Rating']
rate_ = data[rate_pa].groupby(['Payment']).mean()
rate_
```

```python
rate_.plot(kind='barh', grid=True, title="Rating analysis based on Payment", figsize=(10, 5))
plt.show()
```

<b>Looking back at the previous charts it showed that cash mode of payment brought in more revenue than the other modes of payment but, the 'Credit Card' mode of payment recieved more ratings than the other modes of payment, which can denote that the company needs to push the credit card mode of payment more</b>

```python
rate_p = ['Product_line', 'Rating']
rate = data[rate_p].groupby(['Product_line']).mean()
rate
```

```python
rate = rate.sort_values("Rating")
rate.plot(kind='barh', grid=True, title = "Rating analysis based on Product Line", figsize=(10, 5))
```

<b>This chart shows that the 'Food and Beverages' product recieved more rating compared to other products</b>

## RELATIONSHIP ANALYSIS

```python
data.columns
```

```python
u_q = data[['Unit_price', 'Quantity']].corr()
sns.heatmap(u_q, annot=True)
data.plot.scatter(x='Unit_price', y='Quantity')
plt.show()
```

<b>Looking at this plot there is no single correlation between both columns 'Quantity' and 'Unit Price', there is no relationship between both tables</b>

```python
c_t = data[['cogs', 'Total']].corr()
sns.heatmap(c_t, annot=True)
data.plot.scatter(x='cogs', y='Total')
plt.show()
```

<b>Looking at the scatterplot chart and the heatmap there is a high level of correlation between both column 'cogs' and 'Total', both columns have a linear relationship which means as the price of the cogs goes up the total revenue goes up also</b>

```python
u_r = data[['Unit_price', 'Rating']].corr()
sns.heatmap(u_r, annot=True)
data.plot.scatter(x='Unit_price', y='Rating')
plt.show()
```

<b>Looking at both charts there is also no correlation between both columns 'UnitPrice' and 'Rating'; both columns have no relationship</b>
